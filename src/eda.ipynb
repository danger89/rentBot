```python
def remove_outliers(df, cols, r=1.5):
    Q1 = df[cols].quantile(0.25)
    Q3 = df[cols].quantile(0.75)
    IQR = Q3 - Q1

    df = df[~((df[cols] < (Q1 - r * IQR)) | (df[cols] > (Q3 + r * IQR))).any(axis=1)]
    
    return df
```

```python
def corr_matrix(df):
    '''
    Returns a plot (heatmap) of the diagonal correlation matrix for all
    the variables in the dataframe 'df'.

            Parameters:
                    df (dataframe): Dataframe containing the data
    '''

    #Correlation matrix
    df_corr = df.corr()

    #Figure and axis
    fig, ax = plt.subplots(figsize=(15, 10))
    #Mask
    mask = np.triu(np.ones_like(df_corr, dtype=bool))
    #Adjust mask and df
    mask = mask[1:, :-1]
    corr = df_corr.iloc[1:,:-1].copy()
    #Color map
    cmap = sns.diverging_palette(0, 300, 800, 70, as_cmap=True)

    #Plot heatmap
    sns.heatmap(corr, mask=mask, annot=True, fmt=".2f",
               linewidths=5, cmap=cmap, vmin=-1, vmax=1,
               cbar_kws={"shrink": .8}, square=True)

    #Ticks
    yticks = [i.upper() for i in corr.index]
    xticks = [i.upper() for i in corr.columns]
    plt.yticks(plt.yticks()[0], labels=yticks, rotation=0)
    plt.xticks(plt.xticks()[0], labels=xticks, rotation=45)
    #Title
    plt.title('CORRELATION MATRIX', fontsize=25)
    plt.show()
    return


#All variables correlations
corr_matrix(df)
```

```python
def corr_matrix_one_vs_all(df, var):
    '''
    Returns a plot (heatmap) of the correlation matrix with the
    correlations between the variable 'var' and all the other
    variables in the dataframe 'df'.

            Parameters:
                    df (dataframe): Dataframe containing the data
                    var (string): Name of the variable to consider its correlations
    '''

    #Correlation matrix
    df_corr = df.corr()
    #Correlation matrix with var
    df_corr_var = pd.DataFrame(df_corr[var].sort_values(ascending=False))

    #Figure and axis
    fig, ax = plt.subplots(figsize=(15, 10))
    #Mask
    mask_var = np.zeros_like(df_corr_var, dtype=bool)
    mask_var[0] = True
    #Adjust mask and df
    mask_var = mask_var[1:]
    corr_var = df_corr_var.iloc[1:].copy()
    #Color map
    cmap = sns.diverging_palette(0, 300, 800, 70, as_cmap=True)

    #Plot heatmap
    sns.heatmap(corr_var, mask=mask_var, annot=True, fmt=".2f",
               linewidths=5, cmap=cmap, vmin=-1, vmax=1,
               cbar_kws={"shrink": .8}, square=True)

    #Ticks
    yticks = [i.upper() for i in corr_var.index]
    xticks = [i.upper() for i in corr_var.columns]
    plt.yticks(plt.yticks()[0], labels=yticks, rotation=0)
    plt.xticks(plt.xticks()[0], labels=xticks, rotation=0)
    #Title
    plt.title(f'FEATURES CORRELATING WITH {var.upper()}\n', fontsize=20)
    plt.show()
    return


#Features correlating with Price
corr_matrix_one_vs_all(df, 'Precio (€/mes)')
```
